##Popularity based Recommendation System

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
# Rename the Series itself using the 'name' attribute
num_rating_df.name = 'num_ratings'
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df[popularity_df['num_ratings']>=250].sort_values('Book-Rating',ascending=False).head(100)
```

```python
popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','Book-Rating']]
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200

active_users=x[x].index
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
y[y]
```

```python
popular_books=y[y].index
```

```python
final_ratings=filtered_ratings['Book-Title'].isin(popular_books)
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books)]
```

```python

pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')

```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
      print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
